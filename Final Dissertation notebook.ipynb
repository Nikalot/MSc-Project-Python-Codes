```python
from sklearn.model_selection import KFold

def kfold_validation(X, y, model, k=5):
    """
    Perform k-fold cross-validation on the given data using the specified model.

    Parameters:
        X (array-like): The input features.
        y (array-like): The target variable.
        model: The model to be trained and evaluated.
        k (int): The number of folds for cross-validation.

    Returns:
        scores (list): The evaluation scores for each fold.
    """
    kf = KFold(n_splits=k)
    f1_scores = []
    accuracy_scores = []
    precision_scores = []
    recall_scores = []

    for train_index, test_index in kf.split(X):
        X_train, X_test = X[train_index], X[test_index]
        y_train, y_test = y[train_index], y[test_index]

        model.fit(X_train, y_train)
        score = model.score(X_test, y_test)
        accuracy_scores.append(score)
        pred = model.predict(X_test)
        f1_scores.append(f1_score(y_test,pred))
        precision_scores.append(precision_score(y_test,pred))
        recall_scores.append(recall_score(y_test,pred))

    return f1_scores,accuracy_scores, precision_scores,recall_scores
```

```python
f1_scores,accuracy_scores, precision_scores,recall_scores = kfold_validation(X, y, RandomForestClassifier(), k=5)
```

```python
import matplotlib.pyplot as plt

def plot_feature_importance(df, model):
    # Get feature importances
    importances = model.feature_importances_

    # Create a DataFrame with feature names and importances
    feature_importance_df = pd.DataFrame({'Feature': df.columns, 'Importance': importances})

    # Sort the features by importance in descending order
    feature_importance_df = feature_importance_df.sort_values('Importance', ascending=False)

    # Plot the feature importances
    plt.figure(figsize=(10, 6))
    plt.bar(feature_importance_df['Feature'], feature_importance_df['Importance'])
    plt.xticks(rotation='vertical')
    plt.xlabel('Feature')
    plt.ylabel('Importance')
    plt.title('Feature Importance')
    plt.tight_layout()
    plt.show()

    return feature_importance_df
```
